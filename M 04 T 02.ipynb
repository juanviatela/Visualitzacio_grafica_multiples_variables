## Utilización de RegEx para extraer información textual de las columnas. 

Utilizaremos regular expressions (RegEx) para extraer la información de las columnas textuales. Al ser un lenguaje un poco complicado de leer primero escribiré el código y luego lo describire caracter por caracter.

```python
#extracción de columna 'title'
movies[['titulo', 'año']] = movies['title'].str.extract(r'^(.*?)\s\((\d{4})\)$')

#borramos la columna original
movies.drop(columns=["title"], inplace=True)

#revisamos el dataset
movies.head()
```

Hagamos el análisis de esta pieza de código: 

movies[['titulo', 'año']] = movies['title'].str.extract(r'^(.*?)\s\((\d{4})\)$')


movies[['titulo', 'año']]: creamos las dos columnas en donde se alojará los datos que extraemos con el RegEx
movies['title'].str.extract: aplicamos las funciones .str.extract para decirle a python que extraeremos texto de la columna 'title'

RegEx

(r'^(.*?)\s\((\d{4})\)$')

r': nos indica el inicio del RegEx. Se utiliza siempre

Extracción del título: ^(.* ?)\s

- '^': en este caso nos dice que la expresión que estamos buscando está al comienzo del string. 
- (.* ?): aquí tenemos dos cosas: 
    - los paréntesis nos indican que es un grupo de captura. Es decir que de toda el string esta será la información que extraeremos. 
    - punto y estrella nos indican que extraeremos todos los caracteres (.) que se repitan 0 o más veces (* ). El interrogante nos dice que es un grupo no codicioso de captura. Esto es, que intentará cumplir la condición extrayendo el menor número de carácteres. Esto se utiliza para que la expresión no termine extrayendo toda la string. 
'\s': por fuera del primer paréntesis (es decir no hace parte del grupo de captura) usamos \s para indicar un espacio en blanco. Este es el espacio en blanco que separa el título del año y no se extrae. 

Extracción del año: \((\d{4})\)signopesos

-  '\ ( ':  el slash invertido nos indica que el carácter a su derecha (el paréntesis) no se está usando como expresión de captura sino en su sentido literal. Es decir nos indica que el paréntesis significa literalmente paréntesis.
- '(\d{4})' Aquí tenemos un grupo de captura codicioso, es decir que extraera lo que más pueda para cumplir su condición. Primero tenemos \d que quiere decir que extraera solamente dígitos, {4} es una metaexpresión de repetición que nos dice que extraerá digitos que se repiten cuatro veces (el formato estandar como se expresa un año". 
- '\)': hacemos literal el paréntesis de cierra por fuera del grupo de captura. Esto quiere decir que los paréntesis no serán capturados sino la información que está dentro de ellos, es decir los cuatro dígitos del año. 
- "signo pesos" : metacarácter que nos indica que el grupo de captura se ubica al final de la string. 

Hay que tener en cuenta que las RegEx siempre se escriben dentro de comillas luego de la r. 

Nota al pie: como muchos de los carácteres utilizados en la RegEx sirven como indicadores de formato de Markdown este párrafo será díficil de leer. Espero que lo entiendas. 

```python
#extracción de la columna 'genre'
movies["genero"] = movies['genres'].str.extract(r'^([^|]+)')

#borramos la columna original
movies.drop(columns=["genres"], inplace=True)

#revisamos el dataset
movies.head()
```

Analicemos el RegEx para extraer el primer genero de la columna 'genres'

(r'^([^|]+)')

- 'r': indicador de comienzo de la RegEx
- '^': ubicamos la RegEx a comienzo del string
- '()': los parentesis indican la creación de un grupo de captura, es decir que todo lo que esté dentro del paréntesis será capturado. Como no aparece el signo pesos tenemos un grupo de captura codicioso, es decir que capturará lo máximo posible para cumplir la condición. 
- '^': en el contexto del grupo de captura '^' significa negativo. En decir: '^|' significa que queremos capturar cualquier cosa que no sea un '|' del string. Se pone entre corchetes rectos para establecer una relación entre los dos metacaracteres. 
- '+': el signo más es un cuantificador que nos indica que queremos que la expresión que vamos a capturar se repita una o más veces.  

En síntesis esta expresión nos está indicando que quiere capturar todo lo que no sea un '|' desde el principio de la cadena hasta la aparición del primer '|'.

```python
from matplotlib.ticker import FuncFormatter

#configuramos el estilo del plot
sns.set_theme(style="darkgrid")

#determinamos el valor del lienzo utilizando el plt.figure(figsize)
plt.figure(figsize=(8, 6))

# Creamos una función de porcentaje para que el formato de los ticks de x sea porcentaje
def formato_porcentaje(valor, _):
    return f'{valor * 100:.2f}%'

# Diseñamos el barplot
sns.barplot(x=genero_pct.values, y=genero_pct.index, palette='RdYlBu')

#Utilizando la función .gca (get the current axes) llamamos al ejex para correr la función 
plt.gca().xaxis.set_major_formatter(FuncFormatter(formato_porcentaje))

#configuramos títulos y labels
plt.title('Distribución de Películas por Género')
plt.ylabel('Género')

#Pintar el gráfico
plt.show()
```

Para finalizar haremos gráficos comparativos para poder contrastar las tendencias de los 5 géneros en el tiempo. 

```python
#creamos una función para crear datasets basados en genero y año y luego plotear dichos datasets apartir del index 
#y el value resultantes
def plot_genero_por_año(data, genero):
    pelis_genero = data[data['genero'] == genero]
    total_pelis_genero = pelis_genero['año'].value_counts().sort_index()

    plt.plot(total_pelis_genero.index, total_pelis_genero.values, label=genero)

#creamos una lista con los géneros escogidos.     
generos_interes = ['Drama', 'Comedy', 'Action', 'Horror', 'Adventure']

#configuramos el estilo del plot
sns.set_theme(style="darkgrid")

#determinamos el valor del lienzo utilizando el plt.figure(figsize)
plt.figure(figsize=(8, 6))

#creamos un bucle para que la función itere por cada uno de los 5 géneros. 
for genero in generos_interes:
    plot_genero_por_año(moviespequeño, genero)

#configuramos títulos, labeles y legend
plt.title('Número de Películas por Género y Año')
plt.legend()

#pintamos el gráfico
plt.show()
```

En la gráfica podemos observar que 2 de los 5 (drama y comedy) géneros trazan la misma línea ascendente que la tendencia general. Son los mismos géneros que tienen el mayor número de películas. El género action presenta una tendencia intermedia -tiene un ascenso a comienzos de la década de los 90 pero no tan pronunciado como los dos anteriores. Por el contrario Horror y Adventure presentan una tendencia más regular. La producción de películas durante los años analizados de esos dos géneros se mantiene estable. Para poder observar más detalladamente el comportamiento de cada uno de los géneros haremos un subplots

```python
#volvemos a definir nuestra función plot_genero_por_año pero esta vez añadiremos el argumento ax para plotear 
#cada uno de los géneros

def plot_genero_por_año(data, genero, ax):
    pelis_genero = data[data['genero'] == genero]
    total_pelis_genero = pelis_genero['año'].value_counts().sort_index()

    #dentro de la función creamos el ax.plot, definimos estilo y  configuramos títulos y labels. 
    sns.set_theme(style="darkgrid")
    ax.plot(total_pelis_genero.index, total_pelis_genero.values)
    ax.set_title(f'Número de Películas por Año - {genero}')

#creamos fig y axes para crear los sublots. El número de líneas corresponderá al len de generos_interes, 
#una sola columna, compartiran eje x (pues todos se basan en la misma línea de tiempo. )
fig, axes = plt.subplots(nrows=len(generos_interes), ncols=1, figsize=(12, 8), sharex=True)

#creamos el loop for por cada uno de los elementos que habitan la lista generos_interes. Con la función enumerate
#obtenemos tanto el índice como el valor del elemento en la lista. El indice lo asignamos a axes y el genero a genero
for i, genero in enumerate(generos_interes):
    plot_genero_por_año(moviespequeño, genero, axes[i])

#ajusgamos el gráfico para que los titulos no se sobrepongan
plt.tight_layout()

#pintamos los gráficos
plt.show()
```

Este gráfico nos permite ver las especificidades de cada tendencia. Podemos observar que tanto Drama como Comedy presentan tendencias ascendentes desde comienzos de la década de los noventa pero la de Drama parece más estables.También comparten el bache de 1997. 

Acción presenta la misma tendencia pero en una escala menor. 

Los comportamientos de Horror y Adventure parecen mucho menos regulares aunque en Adventure si podemos detectar una tendencia ascendente desde 1992 que llega a su cúspide en 1995 para luego desplomarse. Recupera la tendencia ascendente en 1999.
